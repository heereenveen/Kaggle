# Credit Card Fraud Detection

## Load Data

```python
import kagglehub # type: ignore

path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv(f"{path}/creditcard.csv")
df.head()
```

## Simple EDA

```python
df.info()
```

```python
df.describe().T
```

```python
print("Target values distribution:\n", df['Class'].value_counts())
```

## Data preparation

```python
X = df.drop(labels=['Class'], axis=1)
y = df['Class'].copy()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

first option: oversampling our dataset

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(sampling_strategy="not majority")
X_train_ros, y_train_ros = ros.fit_resample(X_train, y_train)

y_train_ros.value_counts()
```

second option: using SMOTE for dataset

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy="not majority")
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

y_train_smote.value_counts()
```

## Training and Evaluating Decision Tree Classifier model from scikit-learn

```python
from sklearn.tree import DecisionTreeClassifier

dtc_oversampled = DecisionTreeClassifier(random_state=42)
```

Train our model on oversampled dataset and check classification report as result

```python
from sklearn.metrics import classification_report

dtc_oversampled.fit(X_train_ros, y_train_ros)
predicts = dtc_oversampled.predict(X_test)
print(classification_report(y_test, predicts))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, predicts, labels=dtc_oversampled.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dtc_oversampled.classes_)
disp.plot()
plt.show()
```

training and evaluating decision tree on smote dataset

```python
dtc_smote = DecisionTreeClassifier(random_state=42)
dtc_smote.fit(X_train_smote, y_train_smote)
predicts = dtc_smote.predict(X_test)
print(classification_report(y_test, predicts))
```

```python
cm = confusion_matrix(y_test, predicts, labels=dtc_smote.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dtc_smote.classes_)
disp.plot()
plt.show()
```

Check our model on not oversampled dataset

```python
dtc_base = DecisionTreeClassifier(random_state=42)
dtc_base.fit(X_train, y_train)
predicts = dtc_base.predict(X_test)
print(classification_report(y_test, predicts))
```

```python
cm = confusion_matrix(y_test, predicts, labels=dtc_base.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dtc_base.classes_)
disp.plot()
plt.show()
```

Lets evaluate our model by cross-validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(dtc_base, X, y, cv=5, scoring="recall")
scores
```
